```python
#Write a Python program to find the first 20 non-even prime natural numbers.
s = 0
e = 75
for i in range(s,e):
    for j in range(2,i):
        if i%j == 0:
            break
    else:
        print(i)
```

```python
#Write a Python program to implement 15 functions of string.

#1. Capitalize(): The capitalize() method returns a string where the first character is upper case
str = "hello! welcom to AL/ML batch 2"
l = str.capitalize()
print(l)
print("--------------------------------------------------------------------------")

#2. casefold(): Converts string into lower case
str = "DHANVVI KANABAR"
l = str.casefold()
print(l)
print("--------------------------------------------------------------------------")

#3. count(): Returns the number of times a specified value occurs in a string
str = "He loves apples, apples are actually good for health!!"
l = str.count("apples")
print(l)
print("--------------------------------------------------------------------------")

#4.split(): Splits the string at the specified separator, and returns a list
str = "hello! welcome to AL/ML batch 2"
l = str.split()
print(l)
print("--------------------------------------------------------------------------")

#5.upper(): Converts a string into upper case
str = "hello! welcome to ai/ml batch 2"
l = str.upper()
print(l)
print("--------------------------------------------------------------------------")

#6.title(): Converts the first character of each word to upper case
str = "python gets more interesting as we dive deep into its libraries"
l = str.title()
print(l)
print("--------------------------------------------------------------------------")

#7. translate(): Returns a translated string
#use a dictionary with ascii codes to replace 83 (S) with 80 (P)
my_dict = {83 : 80}
str = "Hello Samantha!!"
print(str.translate(my_dict))
print("--------------------------------------------------------------------------")

#8. rindex(): Searches the string for a specified value and returns the last position of where it was found
str = "we create a new Class using the Class statement and the name of the Class"
l = str.rindex("Class")
print(l)
print("--------------------------------------------------------------------------")

#9. lower(): Converts a string into lower case
str = "PYTHON IS INTERESTING"
l = str.lower()
print(l)
print("--------------------------------------------------------------------------")

#10. swapcase(): Swaps cases, lower case becomes upper case and vice versa
str = "PyThOn Is InTeReStInG"
l = str.swapcase()
print(l)
print("--------------------------------------------------------------------------")

#11.find(): Searches the string for a specified value and returns the position of where it was found
#The find() method finds the first occurrence of the specified value.
str = "we create a new Class using the Class statement and the name of the Class"
l = str.find("Class")
print(l)
print("--------------------------------------------------------------------------")

#12. join(): Joins the elements of an iterable to the end of the string
t = ("LetsUpgrade","AI/ML")
n = " ##@*@## ".join(t)
print(n)
print("--------------------------------------------------------------------------")

#13.partition()	Returns a tuple where the string is parted into three parts
str = "I can eat cakes all day long"
n = str.partition("cakes")
print(n)
print("--------------------------------------------------------------------------")

#14. index: Searches the string for a specified value and returns the position of where it was found
str = "I can eat cakes all day long"
n = str.index("cakes")
print(n)

#15. center()	Returns a centered string
t = ("----------------------------------------------------LetsUpgrade-------------------------------------------------------")
n = t.center(70)
print(n)
```

```python
#Write a Python program to check if the given string is a Palindrome or Anagram or None of them. Display the message accordingly to the user

def palindromeCheck(str):
    return str == str[::-1]
#driver code
str = input("Enter a string to chrck whether it is palindrome or not: ")
check = palindromeCheck(str)

if check:
    print("It is palindrome")
else:
    print("Not a palindrome!!")
```
